**Filtrado de datos**

En base al problema debo de primero saber si tengo un docente con titulo de ECONOMISTA dentro del dataset, esto gracias a un conteo de valores unicos en la variable "nombre_titulo" del dataset.

```python
#Para revisar la titularidad especifica
unique_counts = data['nombre_titulo'].value_counts()
print(unique_counts)
```

Para resolver el problema se debe de filtrar la data en base a lo visto en el problema planteado, el nombre de la universidad, el titulo del docente, fecha inicio del periodo donde se rigio la asignatura, sexo del docente y su autoridad academica. Con la data filtrada, tan solo es visualizar los posibles involucrados de la situacion a resolver.

```python
# Suponiendo que "titularidad" y "periodoacademico" son nombres de columnas en tu DataFrame
filtered_data = data[ (data['nombre_ies_internacional'] == 'UNIVERSIDAD CATOLICA DE SANTIAGO DE GUAYAQUIL') & 
                     (data['nombre_titulo'] == 'ECONOMISTA') & 
                     (data['fecha_inicio_periodo'] == 2015) & 
                     (data['sexo'] == 'HOMBRE') & 
                     (data['autoridad_academica'] == 'VICERRECTOR(A) O EQUIVALENTE')]

# Imprimir las primeras filas del DataFrame filtrado para verificar
print(filtered_data.shape)
filtered_data.head()
```

**Con estos datos filtrados, ahora se sabe mas especificamente a cual docente consultar sobre la fuga de informacion personal para llevar a una corte justa**
